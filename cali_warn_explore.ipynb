## Filter by Orange County

```python
pd.options.display.max_rows = 999
orange_county = cali_raw[cali_raw['County'] == 'Orange County']
orange_county.head(20)
```

```python
orange_county['Year'] = orange_county['Effective_Date'].str[-4:]
orange_county.head(20)
```

```python
orange_county_2020 = orange_county[orange_county['Year'] == '2020']
orange_county_2020
```

```python
orange_county_2020['City'].value_counts()
```

```python
layoffs_by_city = orange_county_2020.groupby(['City','Year']).agg('sum')
```

```python
layoffs_by_city.sort_values(by='No_employees', ascending=False)
```

```python
layoffs_by_city['No_employees'].sum()
```

```python
orange_county_2020['Company'].unique()
```

```python
layoffs_by_company = orange_county_2020.groupby(['Company','Year']).agg('sum')
```

```python
layoffs_by_company.sort_values(by='No_employees', ascending=False)
```

Should Categorize & then sum up the numbers.

```python
closures = ['Closure Permanent', 'Closure Temporary']
layoffs = ['Layoff Permanent', 'Layoff Temporary', 'Layoff Unknown at this time']
```

```python
oc_2020_layoffs = orange_county_2020.loc[orange_county_2020['Layoff_Closure'].isin(layoffs)]
oc_2020_layoffs.head(5)
```

```python
oc_2020_closures = orange_county_2020.loc[orange_county_2020['Layoff_Closure'].isin(closures)]
oc_2020_closures.head(5)
```

```python
layoffs_2020 = oc_2020_layoffs.groupby(['City','Year']).agg('sum')
layoffs_2020.sort_values('No_employees', ascending=False)
```

```python
closures_2020 = oc_2020_closures.groupby(['City','Year']).agg('sum')
closures_2020.sort_values('No_employees', ascending=False)
```

```python
# Total Layoffs 2020 - thus far.
layoffs_2020['No_employees'].sum()
```

```python
# Total Closures 2020 - thus far.
oc_2020_closures['No_employees'].sum()
```
