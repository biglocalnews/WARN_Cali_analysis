## How we got this California Historical Data - Big Local News

We manually downloaded the PDFs from the california WARN notices site: https://edd.ca.gov/Jobs_and_Training/Layoff_Services_WARN.htm

We only grabbed the historical data which (on 10/22/20) resulted in the downloading of 6 PDF files with data ranging from 2014-2020.

After downloading the data, we used Tabula to convert each PDF to a csv file.
The exported csv files are the same that we are importing below in order to create one large historical file.

See `cali_hist_data` directory for all the data.

```python
import pandas as pd
```

```python
cali_19_20 = pd.read_csv('cali_hist_data/cali_2019_20.csv')
cali_18_19 = pd.read_csv('cali_hist_data/cali_2018_19.csv')
cali_17_18 = pd.read_csv('cali_hist_data/cali_2017_18.csv')
cali_16_17 = pd.read_csv('cali_hist_data/cali_2016_17.csv')
cali_15_16 = pd.read_csv('cali_hist_data/cali_2015_16.csv')
cali_14_15 = pd.read_csv('cali_hist_data/cali_2014_15.csv')
```

## Now for a Teeny bit of Data Cleaning

I exported the `cali_hist_data` file and saw that the format was wonky because there are some new line and new row characters within the company names. Also, the `Employees` column for the `cali_2019_20` dataframe should be converted to integers so that the data in the column is of the same type.

```python
cali_hist_data['Company'] = cali_hist_data['Company'].str.replace('\\r', '')
cali_hist_data['Company'] = cali_hist_data['Company'].str.replace('\\n', '')
cali_hist_data['City'] = cali_hist_data['City'].str.replace('\\r', '')
cali_hist_data['City'] = cali_hist_data['City'].str.replace('\\n', '')
cali_hist_data['Layoff/Closure'] = cali_hist_data['Layoff/Closure'].str.replace('\\r', '')
cali_hist_data['Layoff/Closure'] = cali_hist_data['Layoff/Closure'].str.replace('\\n', '')
```

```python
cali_hist_data.to_csv('cali_hist_data/cali_warn_hist.csv', index=False)
```
