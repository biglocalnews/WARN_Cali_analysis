```python
import pandas as pd
```

### MEDIAN GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME IN THE PAST 5 Years (2018 - 5 year estimate)

Link to ACS site that holds this table - in case we need to revisit for more data: https://data.census.gov/cedsci/table?q=rent%20as%20percentage%20of%20income&g=0400000US06.050000&tid=ACSDT5Y2018.B25071&tp=true&hidePreview=true

```python
ACS_2018_5year_rent_median = pd.read_csv('data/ACS_2018_5year_median_gross_rent_percent_of_income.csv')
housing_authorities = pd.read_csv('data/Housingauthorities.csv')
```

```python
ACS_2018_5year_rent_median.head()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
housing_authorities.head(35)
```

```python
ACS_2018_5year_rent_median['County Fips Code'] = ACS_2018_5year_rent_median['GEO_ID'].str[-5:]
ACS_2018_5year_rent_median = ACS_2018_5year_rent_median.drop(0)
ACS_2018_5year_rent_median.rename(columns = {'B25071_001E':'Median Gross Rent as % of Household income', 'B25071_001M':'Margin of Error'}, inplace=True)
ACS_2018_5year_rent_median.head()
```

```python
housing_authorities.head(5)
```

```python
cali_housing_authorities = housing_authorities.loc[housing_authorities['STATE2KX'] == 6]
cali_housing_authorities.head(5)
len(cali_housing_authorities)
```

```python
cali_housing_authorities['COUNTY_LEVEL'] = pd.to_numeric(cali_housing_authorities['COUNTY_LEVEL'], downcast='integer')
```

```python
cali_housing_authorities.head()
```

```python
cali_housing_authorities['COUNTY_LEVEL'].dtype
cali_housing_authorities['COUNTY_LEVEL'] = cali_housing_authorities['COUNTY_LEVEL'].astype('str')
cali_housing_authorities['COUNTY_LEVEL'] = cali_housing_authorities['COUNTY_LEVEL'].str.zfill(5)
cali_housing_authorities.head(5)
len(cali_housing_authorities)
```

```python
ACS_rent_median_housing_authorities = pd.merge(cali_housing_authorities, ACS_2018_5year_rent_median, how='left', left_on='COUNTY_LEVEL', right_on='County Fips Code')
ACS_rent_median_housing_authorities.head()
```

```python
ACS_2018_5year_rent_median.sort_values(by=['Median Gross Rent as % of Household income'], ascending=False).head()
```

```python
ACS_2018_rent_burden = pd.read_csv('data/ACS_2018_5year_rentburden.csv')
ACS_2018_rent_burden.head()
```

```python
ACS_2018_rent_burden['COUNTY FIPS'] = ACS_2018_rent_burden['GEO_ID'].str[-5:]
ACS_2018_rent_burden = ACS_2018_rent_burden.drop(0)
ACS_2018_rent_burden.head()
```

```python
full_table = pd.merge(ACS_rent_median_housing_authorities, ACS_2018_rent_burden, how='left', left_on='COUNTY_LEVEL', right_on='COUNTY FIPS')
full_table.sort_values(by='Rent Burden Percent', ascending=False).head()
```

```python
full_table.to_csv('data/rent_burden_households_and_median_gross_rent_percent.csv')
```
