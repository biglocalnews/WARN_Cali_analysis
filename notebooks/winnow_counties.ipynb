```python
import pandas as pd
```

```python
## Add link where you got this data.

ACS_2018 = pd.read_csv('../data/rent_burden/ACS_2018_5year_est.csv')
housing_authorities = pd.read_csv('../data/rent_burden/Housingauthorities.csv')
ACS_2018.head()
```

```python
ACS_2018_fifty_percent = ACS_2018[['GEO_ID', 'NAME', 'B25070_001E', 'B25070_001M', 'B25070_010E', 'B25070_010M']]
ACS_2018_fifty_percent.head()
```

```python
ACS_2018_fifty_percent.rename(columns = {'B25070_001E':'Estimate Total', 'B25070_001M':'Margin of Error', 'B25070_010E':'50 Percent or More', 'B25070_010M':'Margin of Error for 50 percent or more'}, inplace=True)
ACS_2018_fifty_percent.head()
```

```python
ACS_2018_fifty_percent = ACS_2018_fifty_percent[1:]
ACS_2018_fifty_percent.head()
```

```python
ACS_2018_fifty_percent['50 Percent or More'] = ACS_2018_fifty_percent['50 Percent or More'].astype('int64')
ACS_2018_fifty_percent['Estimate Total'] = ACS_2018_fifty_percent['Estimate Total'].astype('int64')
ACS_2018_fifty_percent['Rent Burden Percent'] = (ACS_2018_fifty_percent['50 Percent or More']/ACS_2018_fifty_percent['Estimate Total']) * 100
ACS_2018_fifty_percent.head()
```

```python
sorted_ACS_2018 = ACS_2018_fifty_percent.sort_values(by='Rent Burden Percent', ascending=False)
sorted_ACS_2018.head()
```

```python
# sorted_ACS_2018.to_csv('../data/rent_burden/ACS_2018_5year_rentburden.csv')
```

```python
cali_housing_authorities = housing_authorities.loc[housing_authorities['STATE2KX'] == 6]
cali_housing_authorities.head(5)
len(cali_housing_authorities)
```

```python
cali_housing_authorities['COUNTY_LEVEL'] = pd.to_numeric(cali_housing_authorities['COUNTY_LEVEL'], downcast='integer')
cali_housing_authorities.head()
```

```python
cali_housing_authorities['COUNTY_LEVEL'].dtype
cali_housing_authorities['COUNTY_LEVEL'] = cali_housing_authorities['COUNTY_LEVEL'].astype('str')
cali_housing_authorities['COUNTY_LEVEL'] = cali_housing_authorities['COUNTY_LEVEL'].str.zfill(5)
cali_housing_authorities.head(5)
len(cali_housing_authorities)
```

```python
sorted_ACS_2018['COUNTY FIPS'] = sorted_ACS_2018['GEO_ID'].str[-5:]
sorted_ACS_2018.head()
```

```python
full_table = pd.merge(cali_housing_authorities, sorted_ACS_2018, how='left', left_on='COUNTY_LEVEL', right_on='COUNTY FIPS')
full_table.sort_values(by='Rent Burden Percent', ascending=False).head()
```

```python
full_table.to_csv('../data/rent_burden/rent_burden_pha_expanded.csv')
```
