### Number of Companies that Filed Notices

```python
## Creating separate dataframes for 2019 records and 2020 records

cali_2019_info = warn_data[warn_data['Year'] == 2019]
cali_2020_info = warn_data[warn_data['Year'] == 2020]
```

### Layoffs by Type in California overall

```python
## Job loss by layoff type 2019

figures_2019 = cali_2019_info.groupby(['Layoff/Closure clean']).sum()
figures_2019.reset_index()
figures_2019.drop(columns=['Year','Population'], inplace=True)
```

```python
## Job loss by layoff type 2020

figures_2020 = cali_2020_info.groupby(['Layoff/Closure clean']).sum()
figures_2020.reset_index()
figures_2020.drop(columns=['Year','Population'], inplace=True)
```

```python
# Merging the two tables above to get job loss by layoff type for 2020 and 2019

layoffs_20_19 = pd.merge(figures_2020, figures_2019, how='left', left_on="Layoff/Closure clean", right_on="Layoff/Closure clean")
layoffs_20_19.rename(columns={'Employees_x':'Employees Laid Off 2020', 'Employees_y':'Employees Laid Off 2019'}, inplace=True)
layoffs_20_19
```

### Layoffs by Type and County

```python
# Getting layoff records grouped by type and county for 2019
# You can see all of Alameda County layoffs by type
# List continues into all of other California counties but gets truncated due to length

layoff_counties_2019 = cali_2019_info[['County','Employees','Layoff/Closure clean']]
layoff_counties_2019 = layoff_counties_2019.groupby(['County','Layoff/Closure clean']).sum()
layoff_counties_2019.reset_index('Layoff/Closure clean')
```

```python
# Getting layoff records grouped by type and county for 2020
# You can see all of Alameda County layoffs by type
# List continues into all of other California counties but gets truncated due to length

layoff_counties_2020 = cali_2020_info[['County','Employees','Layoff/Closure clean']]
layoff_counties_2020 = layoff_counties_2020.groupby(['County', 'Layoff/Closure clean']).sum()
layoff_counties_2020.reset_index('Layoff/Closure clean')
```

```python
# Merged layoff records grouped by type and county for 2020 and 2019

layoffs_merged = pd.merge(layoff_counties_2020, layoff_counties_2019, how='left', on=['County','Layoff/Closure clean'])
layoffs_merged.rename(columns={'Employees_x':'2020 Employees','Employees_y':'2019 Employees'})
```

### High Rent Burden County Layoffs

```python
# Created list of Counties we investigated
# Most of these counties have a significant % of households paying more than 50% of their income in rent

counties_list = ['Butte County', 'Santa Cruz County', 'Yolo County', 'Orange County', 'Los Angeles County', 'Humboldt County', 'Fresno County']
```

```python
# Create dataframes from 2020 and 2019 data
# These new dataframes only include the counties in the list above

high_rent_burden_20 = cali_2020_info[cali_2020_info['County'].isin(counties_list)]
high_rent_burden_19 = cali_2019_info[cali_2019_info['County'].isin(counties_list)]
```

```python
# Hrb stands for high rent burden
# Groups the 2020 dataframe created above by layoff type + county and sums the layoffs

layoff_counties_hrb_2020 = high_rent_burden_20[['County','Employees','Layoff/Closure clean']]
layoff_counties_hrb_2020 = layoff_counties_hrb_2020.groupby(['County', 'Layoff/Closure clean']).agg('sum')
layoff_counties_hrb_2020.reset_index('Layoff/Closure clean')
```

```python
# Groups the 2019 dataframe created above by layoff type + county and sums the layoffs

layoff_counties_hrb_2019 = high_rent_burden_19[['County','Employees','Layoff/Closure clean']]
layoff_counties_hrb_2019 = layoff_counties_hrb_2019.groupby(['County', 'Layoff/Closure clean']).agg('sum')
layoff_counties_hrb_2019.reset_index('Layoff/Closure clean')
```

```python
# Merges the two dataframes above
# You can see layoffs by type and county for 2020 and 2019, instead of separately
# This table contains the counties we were most interested in

layoff_counties_hrb = pd.merge(layoff_counties_hrb_2020, layoff_counties_hrb_2019, how='left', on=['County','Layoff/Closure clean'])
layoff_counties_hrb.rename(columns={'Employees_x':'2020 Employees','Employees_y':'2019 Employees'}, inplace=True)
layoff_counties_hrb
```

### Layoff Rates for High Rent Burden Counties

```python
# Creates a dataframe from population data for each county

population = warn_data[['County','Population']]
population = population.groupby('County').max()
population
```

```python
# We use the layoff_counties_hrb dataframe from above
# We use the population dataframe from above
# We calculate the rate of job loss per 1000 people for 2019 and 2020 based on # of layoffs and county population

rates = pd.merge(layoff_counties_hrb.reset_index('Layoff/Closure clean'), population, how='left', on=['County'])
rates['Rate Laid Off 2020'] = (rates['2020 Employees']/rates['Population']) * 1000
rates['Rate Laid Off 2019'] = (rates['2019 Employees']/rates['Population']) * 1000
rates = rates[['Layoff/Closure clean','2020 Employees','Rate Laid Off 2020','2019 Employees','Rate Laid Off 2019','Population']]
rates
```

### Layoff Rates for High Rent Burden Counties, permanent and temporary layoffs only

```python
# This is a list of the two layoff types we care about most

layoffs = ['layoff permanent', 'layoff temporary']
```

```python
# We make a new dataframe based on the rates dataframe
# This dataframe only includes records that have a layoff type of 'layoff permanent' or 'layoff temporary'

temp_perm_layoffs_only = rates[rates['Layoff/Closure clean'].isin(layoffs)]
temp_perm_layoffs_only
```
